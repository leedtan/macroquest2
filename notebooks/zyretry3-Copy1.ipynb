```python
def serialize(msg):
    hexmsg = ''
    length = len(msg)
    hexend = hex(length)[2:]
    lenhex = len(hexend)
    extrahexlen = 8 - lenhex
    for i in range(8):
        if i % 2 == 0:
            hexmsg += '\x'
        if i < extrahexlen:
            hexmsg += '0'
        else:
            hexmsg += hexend[8-extrahexlen]
```

```python
def whisperdbg(self, peer, msg_p):
    """Send message to single peer, specified as a UUID string
    Destroys message after sending"""
    self.actor.send_unicode("WHISPER", flags=zmq.SNDMORE)
    pdb.set_trace()
    self.actor.send(peer.bytes, flags=zmq.SNDMORE)
    if isinstance(msg_p, list):
        self.actor.send_multipart(msg_p)
    else:
        self.actor.send(msg_p)
```
